```python
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import joblib

# Load data
df = pd.read_csv("heart_failure_clinical_records.csv")
X = df.drop("DEATH_EVENT", axis=1)
y = df["DEATH_EVENT"]

# Train model
model = RandomForestClassifier()
model.fit(X, y)

# Save model
joblib.dump(model, "hfp.pkl")
```
